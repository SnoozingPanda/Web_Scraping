```python
def crawl(url_short):
    url = "https://www.allocine.fr" + url_short
    html = requests.get(url)
    soup = BeautifulSoup(html.text, 'html.parser')
    detail = soup.find("div", {"class" : "meta-body-item meta-body-info"})
    regex_output = re.findall("(\d+)(h )(\d+)(min)", str(detail))[0]
    return int(regex_output[0]) * 60 + int(regex_output[2])
```

```python
def crawl(url_short):
    url = "https://www.allocine.fr" + url_short
    html = requests.get(url)
    soup = BeautifulSoup(html.text, 'html.parser')
    detail = soup.find("div", {"class" : "meta-body-item meta-body-info"})
    try:
        regex_output = re.findall("(\d+)(h )(\d+)(min)", str(detail))[0]
        return int(regex_output[0]) * 60 + int(regex_output[2])
    except:
        pass
```

```python
df_url["duration_min"] = df_url["href"].apply(crawl)
df_url["duration_min"].mean()
```

```python
import json

url = "https://www.allocine.fr" + "/film/fichefilm-114782/casting/"
html = requests.get(url)
soup = BeautifulSoup(html.text, 'html.parser')
```

```python
script_file = soup.find('script', {"type" : "application/ld+json"})
json_data = json.loads(script_file.string)
```

```python
# ici je récupère les pages perso des acteurs (pour ceux qui en ont)
# je créé une liste des identifiants des acteurs

urls = []
numero = []

for i in range(len(json_data)):
  urls.append(json_data["actor"][i]["url"])
  numero.append(re.findall('\d+', json_data["actor"][i]["url"]))

print(urls)
print(numero)
```

```python
# généralisation à la liste des id des acteurs
actor_list = []
birth_list = []

for indice, url in enumerate(urls):
    html2 = requests.get(url)
    soup2 = BeautifulSoup(html2.text, 'html.parser')
    actor_list.append(soup2.find('div', {"class" : "meta-body"}).get_text())

for i in actor_list:
    birth_list.append(re.findall(r'\d+', str(actor_list)))

vraie_birth_list = birth_list[1][1::3]
print(vraie_birth_list)
```

```python
import statistics # sinon je n'aurai pas la fonction mean

# je dois transformer les éléments de la liste en int pour le calcul de moyenne
liste_birth_int = []
for indice, date in enumerate(vraie_birth_list):
  x = int(date)
  liste_birth_int.append(x)

mean = statistics.mean(liste_birth_int)
print(mean)
```

```python
age_moyen = 2024 - mean
print(age_moyen)
```

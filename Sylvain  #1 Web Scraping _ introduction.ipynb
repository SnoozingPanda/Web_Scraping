##Crée un dictionnaire vide, puis crée une boucle qui viendra enregistrer dans ce dictionnaire chaque blague en clé et chaque note correspondante en valeur

```python
dico = {}

blague = soup.find_all('div', {"class" : "card"})


for fact in blague:
  fact_text = fact.find("p", {"class": "card-text"}).get_text(strip=True)
  fact_rating = fact.find("span").get_text(strip=True)
  dico[fact_text] = fact_rating

for fact, rating in dico.items():
  print(f"{fact}: {rating}")
```

##Transforme ce dictionnaire en DataFrame de 2 colonnes : blague et note. Il doit avoir 20 lignes : une par blague

```python
blague_list = []
note_list = []

for fact, rating in dico.items():
  blague_list.append(fact)
  note_list.append(rating)

data = {
    "blague" : blague_list,
    "note" : note_list
}

df = pd.DataFrame(data)
```
